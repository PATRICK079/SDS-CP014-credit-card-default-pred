## Dataset Information
---

This research aimed at the case of customers' default payments in Taiwan and compares the predictive accuracy of probability of default among six data mining methods.

From the perspective of risk management, the result of predictive accuracy of the estimated probability of default will be more valuable than the binary result of classification - credible or not credible clients. Because the real probability of default is unknown, this study presented the novel Sorting Smoothing Method to estimate the real probability of default. With the real probability of default as the response variable (Y), and the predictive probability of default as the independent variable (X), the simple linear regression result (Y = A + BX) shows that the forecasting model produced by artificial neural network has the highest coefficient of determination; its regression intercept (A) is close to zero, and regression coefficient (B) to one.

Therefore, among the six data mining techniques, artificial neural network is the only one that can accurately estimate the real probability of default.

# Week 2 Building new features

### Aggregrating Bill and Payment features

```python
# Aggregate bill amounts
df['BILL_AMT_MEAN'] = df[[f'BILL_AMT{i}' for i in range(1, 7)]].mean(axis=1)
df['BILL_AMT_SUM'] = df[[f'BILL_AMT{i}' for i in range(1, 7)]].sum(axis=1)
df['BILL_AMT_STD'] = df[[f'BILL_AMT{i}' for i in range(1, 7)]].std(axis=1)

# Aggregate payment amounts
df['PAY_AMT_MEAN'] = df[[f'PAY_AMT{i}' for i in range(1, 7)]].mean(axis=1)
df['PAY_AMT_SUM'] = df[[f'PAY_AMT{i}' for i in range(1, 7)]].sum(axis=1)
df['PAY_AMT_STD'] = df[[f'PAY_AMT{i}' for i in range(1, 7)]].std(axis=1)

# Display the first few rows of the aggregated columns
df[['BILL_AMT_MEAN', 'BILL_AMT_SUM', 'BILL_AMT_STD', 'PAY_AMT_MEAN', 'PAY_AMT_SUM', 'PAY_AMT_STD']].head()
```

```python
df[['BILL_AMT_MEAN', 'BILL_AMT_SUM', 'PAY_AMT_MEAN', 'PAY_AMT_SUM', 'default payment next month']].corr()
```

```python
df['PAY_TO_BILL_RATIO'] = df['PAY_AMT_SUM'] / (df['BILL_AMT_SUM'] + 1e-10)  # Avoid division by zero
```

```python


# Define the new aggregated features to visualize
agg_features = {
    'BILL_AMT_MEAN': 'Average Bill Amount',
    'BILL_AMT_SUM': 'Total Bill Amount',
    'BILL_AMT_STD': 'Bill Amount Variability',
    'PAY_AMT_MEAN': 'Average Payment Amount',
    'PAY_AMT_SUM': 'Total Payment Amount',
    'PAY_TO_BILL_RATIO': 'Payment-to-Bill Ratio'
}

# Loop through each feature and plot stacked histograms
for feature, label in agg_features.items():
    plt.figure(figsize=(8, 6))
    
    # Create histogram for Non-Defaulters (0)
    sns.histplot(
        data=df[df['default payment next month'] == 0],
        x=feature,
        bins=30,
        color='green',
        label='Non-Default (0)',
        alpha=0.6
    )
    
    # Create histogram for Defaulters (1), stacked on top
    sns.histplot(
        data=df[df['default payment next month'] == 1],
        x=feature,
        bins=30,
        color='red',
        label='Default (1)',
        alpha=0.6
    )
    
    # Titles and labels
    plt.title(f"Stacked Distribution of {label} by Default Status", fontsize=16)
    plt.xlabel(label, fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.legend(title="Default Status", loc='upper right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

#### Rationale for aggregating features (creating new features)

From the correlation matrix, I was able to determine the PAY*, BILL_AMT*, and PAY_AMT* features are strongly correlated to the target variable "default payment next month". However, since there are so many features, aggregating reduces the clutter without sacrificing much of the predictive capability. 

I aggregated the features with the following: 
* Mean = Average value of months. 
* Sum = Total value of months 
* Standard Deviation = Captures flucuation in bills or payments. 

Here's the breakdown of the new features: 

BILL_AMT_MEAN (Mean Bill Amount):  The average bill amount across the six months (BILL_AMT1 to BILL_AMT6).
* Higher values indicate customers with consistently large bills.

Potential Insights:

* Clients with high mean bill amounts might be at greater risk of default, especially if their payments (PAY_AMT_MEAN) are low.
* Conversely, customers with low mean bill amounts may carry less financial risk.

Check for: Correlation with default payment next month to see if larger average bills correlate with a higher likelihood of default.

BILL_AMT_SUM (Total Bill Amount): The total of all bill amounts over six months.
* Higher values indicate customers who owe more in total across the billing period.

Potential Insights:
* Similar to BILL_AMT_MEAN, but this captures cumulative exposure rather than an average. Customers with very high total bills may represent high financial risk if payments are not proportional.

Check for: Compare this feature with PAY_AMT_SUM (total payments) to identify underpayment trends.

BILL_AMT_STD (Standard Deviation of Bill Amounts): Measures the variability in bill amounts over six months.High variability suggests fluctuations in spending, while low variability indicates consistent billing.

Potential Insights:
* High variability might indicate unstable financial behavior, which could increase default risk.
Stable billing patterns (low variability) might correlate with better financial planning.

PAY_AMT_MEAN (Mean Payment Amount): The average payment made across six months (PAY_AMT1 to PAY_AMT6).
Higher values indicate customers making larger, consistent payments.

Potential Insights:
* Customers with high average payments are less likely to default.
* Low values may indicate financial stress or minimal payments, which could signal higher default risk.

PAY_AMT_SUM (Total Payment Amount): The cumulative amount paid across the six months. Helps assess the total repayment behavior relative to total bills.

Potential Insights:
* Compare this feature to BILL_AMT_SUM. If PAY_AMT_SUM is significantly lower, it indicates underpayment over the period.
* High total payments likely reduce default risk.

 PAY_AMT_STD (Standard Deviation of Payment Amounts) = 
Measures the variability in payments over six months. High variability indicates irregular payment behavior, while low variability shows consistency.

Potential Insights:
* Irregular payments may suggest financial instability, increasing the likelihood of default.
* Consistent payment behavior might correlate with reliability.

PAY_TO_BILL_RATIO (Ratio of Total Payments to Total Bills): Captures how much of the total bill was paid over the six months.

Values:
* Close to 1.0: Indicates most or all bills were paid.
* Below 1.0: Indicates underpayment.
* Above 1.0: Indicates overpayment or early payments.

Potential Insights:
* High ratios (>1.0) indicate responsible financial behavior and reduce default risk.
* Low ratios (<1.0) suggest financial stress or inability to meet obligations, increasing default risk.
